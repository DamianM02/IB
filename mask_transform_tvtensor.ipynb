```python
from torchvision.transforms.v2 import Compose, RandomResizedCrop
import torchvision
import torch
import matplotlib.pyplot as plt

# Przykładowy obraz i maska
img = torch.randint(0, 256, (3, 4, 4), dtype=torch.uint8)  # obraz RGB
mask = torch.randint(0, 2, (4, 4), dtype=torch.uint8)  # maska (0 - tło, 1 - obiekt)
mask = torchvision.tv_tensors.Mask(mask.float())

plt.imshow(img.permute(1,2,0))
plt.show()

plt.imshow(mask)
plt.gray()
plt.show()
```

```python
# Kompozycja transformacji
transform = Compose([
    RandomResizedCrop(size=(256, 256), scale=(0.1, 0.5)),
])

# Słownik z obrazem i maską
sample = {"image": img, "mask": mask}

# Transformacja
out_img, out_mask = transform(sample).values()

plt.imshow(out_img.permute(1,2,0))
plt.show()

plt.imshow(out_mask)
plt.gray()
plt.show()
```
